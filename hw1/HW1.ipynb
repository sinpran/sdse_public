Problem 5

```python
#a 
N = 45
mean = 0.165
sigma = 0.022

lower95 = stats.norm.ppf(0.025) * np.sqrt(sigma**2 / N) + mean
upper95 = stats.norm.ppf(0.975) * np.sqrt(sigma**2 / N) + mean

print('a =',[lower95, upper95])
```

```python
#b 
N = 45
mean = 0.165
sigma = 0.022

confidence_level = np.linspace(0, 1, 100)
confidence_interval_size = 2* abs(stats.norm.ppf((1 - confidence_level)/2)) * np.sqrt(sigma**2 / N)
plt.plot(confidence_level, confidence_interval_size)
plt.xlabel('Confidence Level')
plt.ylabel("Confidence Interval Size")
```

```python
#c 
N = np.linspace(1, 100, 100)
confidence_interval_size = 2* abs(stats.norm.ppf(0.025)) * np.sqrt(sigma**2 / N)
plt.plot(N, confidence_interval_size)
plt.xlabel('Sample Size')
plt.ylabel("Confidence Interval Size")
```

```python
#d 
N = 45
mean = 0.165
sigma = 0.022

left = (0.157 - mean) / (np.sqrt(sigma**2 / N))
right = (0.167 - mean) / (np.sqrt(sigma**2 / N))
confidence = stats.norm.cdf(right) - stats.norm.cdf(left)
print(confidence)
```

```python
#e
N = 45
mean = 0.165
sigma = 0.022
x = np.linspace(mean-5*sigma, mean+5*sigma, 100)
y = stats.norm(loc = mean, scale = sigma).pdf(x)
plt.plot(x, y)
x_fill = np.linspace(0.157, 0.167)
y_fill = stats.norm(loc = mean, scale = sigma).pdf(x_fill)
plt.fill_between(x_fill, y_fill)
```
